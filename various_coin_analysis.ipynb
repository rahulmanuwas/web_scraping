```python
# Imports
%matplotlib inline
import pandas as pd
import requests
from bs4 import BeautifulSoup
import hashlib, json, csv
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Library settings
color = sns.color_palette()
pd.options.mode.chained_assignment = None
pd.options.display.max_columns = 50
pd.options.display.max_rows = 50
pd.set_option('display.float_format', lambda x: '%.3f' % x)
```

```python
# Fetching data - API from https://coinmarketcap.com/api/
api = requests.get("https://api.coinmarketcap.com/v1/ticker/")
api_data = pd.read_json(json.dumps(api.json(), separators=(',', ':')))
```

```python
api_data['last_updated'] = pd.to_datetime(api_data['last_updated'],unit='s')
api_data.head()
```

```python
api2 = requests.get("https://www.cryptocompare.com/api/data/coinlist/")
```

```python
api2_data = pd.read_json(json.dumps(api2.json()["Data"], separators=(',', ':'))).transpose()
```

```python
api2_data
```

```python
#new_df = pd.merge(A_df, B_df,  how='left', left_on=['A_c1','c2'], right_on = ['B_c1','c2'])
merged_data = pd.merge(api_data, api2_data,  how='left', left_on=['symbol'], \
                  right_on = ['Name']).reset_index(inplace=True)
```
